```python
import pandas as pd
from sklearn import preprocessing
```

```python
raw_df = pd.read_excel("Input/full_long_table.xlsx", index_col = "Sample_name")

le = preprocessing.LabelEncoder()
raw_df["Sequence"] = le.fit_transform(raw_df["Sequence"])

raw_df.head()
```

```python
STRS = list(set(raw_df["STR_mark"]))
# print(STRS)
SAMPLES = list(set(raw_df.index))
# print(SAMPLES)

raw_dict = dict()
df_dict = dict()

for STR in STRS:
    raw_dict[STR] = dict()
    df_dict[STR] = dict()
    for sample in SAMPLES:
        raw_dict[STR][sample] = 0
        df_dict[STR][sample] = "None"

# print(df_dict)

# Template dataframe
# template_df = pd.DataFrame(columns=STRS, index=SAMPLES)
# template_df
```

```python
for index, row in raw_df.iterrows():
    al_freq = row["AlleleFrequency"]
    if raw_dict[STR][sample] < al_freq:
        STR = row["STR_mark"]
        sample = index
        seq = row["Sequence"]
        raw_dict[STR][sample] = al_freq
        df_dict[STR][sample] = seq   
```

```python
df = pd.DataFrame.from_dict(df_dict)
df
```
